```python
import numpy as np
import pandas as pd
import glob

DATA_DIR = "../data"

# 1. Load the features dataframe you already processed and saved
df_moltbook = pd.read_csv(f"{DATA_DIR}/moltbook_features_1.csv")

# 2. Load and combine the embedding chunks from your checkpoints
checkpoint_dir = f"{DATA_DIR}/checkpoints_moltbook"
all_files = sorted(glob.glob(f"{checkpoint_dir}/embeddings_part_*.npy"), 
                   key=lambda x: int(x.split('_part_')[1].split('.npy')[0]))
moltbook_embeddings = np.vstack([np.load(f) for f in all_files])

# 3. Define your exact save function (omitting the unnecessary .to_pandas())
def save_unsplit_features(df, embeddings, output_dir="../data", dataset_prefix="data"):
    # Horizontally stack the comment_existence feature with the embeddings for the ENTIRE dataset
    X_full = np.hstack([
        df[['comment_existence']].values,
        embeddings
    ])
    
    # Save the combined full features to a single numpy file
    features_filename = f"{output_dir}/{dataset_prefix}_features_full.npy"
    np.save(features_filename, X_full)
    print(f"Saved full un-split features shape: {X_full.shape} to {features_filename}")
    
    return df

# 4. Execute the save function without splitting
df_final_results = save_unsplit_features(
    df_moltbook, 
    moltbook_embeddings, 
    output_dir=DATA_DIR, 
    dataset_prefix="moltbook"
)

# 5. Save the full metadata reference frame to a pickle
df_final_results.to_pickle(f"{DATA_DIR}/moltbook_metadata.pkl")
print("Full un-split pipeline executed successfully!")
```
