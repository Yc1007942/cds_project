```python
conn = sqlite3.connect(f"{DATA_DIR}/moltbook_upload.db")


# Load all posts into a DataFrame
posts_df = pd.read_sql_query("SELECT * FROM posts", conn)

# Save to CSV
posts_df.to_csv(f"{DATA_DIR}/moltbook_all_posts.csv", index=False)
print(f"Saved {len(posts_df)} posts to all_posts.csv")

conn.close()
```

```python
posts_with_agents = pd.read_sql_query("""
    SELECT p.*, a.karma as author_karma, a.follower_count as author_followers
    FROM posts p
    LEFT JOIN agents a ON p.author_id = a.id
""", conn)
posts_with_agents.to_csv("all_posts_with_agents.csv", index=False)
```
